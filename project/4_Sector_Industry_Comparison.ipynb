# Question To Answer
- How do different sectors and industries perform in terms of returns and risk (volatility) over time?

## Objective
- Compare returns and volatility across different sectors and industries

## Methodology
- Calculate Sector and Industry Returns
- Calculate Sector-level volatility
- Plot Sector Returns and Volatility

```python
# Loading Libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading merged datasets
df_merged = pd.read_csv(r'C:\Users\hp\OneDrive\Desktop\Data_Analyst\S&P_500\sp500_data\cleaned_merged_stock_data.csv')
```

```python
# Calculate Sector and Industry Returns
sector_returns = df_merged.groupby(['Date', 'Sector'])['Daily_Return'].mean().unstack()
industry_returns = df_merged.groupby(['Date', 'Industry'])['Daily_Return'].mean().unstack()

# Calculate Sector-level volatility
sector_volatility = df_merged.groupby(['Date', 'Sector'])['30D_Rolling_Volatility'].mean().unstack()
```

```python
# Plot Sector Returns and Volatility
fig, ax = plt.subplots(2, 1, figsize=(12, 10))

# Plot Sector Returns
sector_returns.plot(ax=ax[0], legend=False)  # Temporarily hide legend for this plot
ax[0].set_title('Average Daily Returns by Sector')
ax[0].set_ylabel('Average Return')
ax[0].grid()

# Add legend outside for Sector Returns
handles, labels = ax[0].get_legend_handles_labels()
fig.legend(handles, labels, loc='center left', bbox_to_anchor=(1.05, 0.5), title='Sectors')

# Plot Sector Volatility
sector_volatility.plot(ax=ax[1], colormap='tab20', legend=False)  # Temporarily hide legend for this plot
ax[1].set_title('Rolling Volatility by Sector')
ax[1].set_ylabel('30-Day Rolling Volatility')
ax[1].grid()
plt.xlabel('Date')

# Adjust layout and show
plt.tight_layout(rect=[0, 0, 0.85, 1])  # Leave space for the legend on the right
plt.show()
```
